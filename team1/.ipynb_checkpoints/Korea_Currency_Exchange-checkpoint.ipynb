```python
import matplotlib.pyplot as plt
import pandas as pd
# import numpy as np
data = pd.read_csv('Korea_CER.csv')
```

```python
# Data Cleaning
# Change the column name to something that is more relevant
# Replace non integer values with zero
data.columns = ['date', 'currency']
data.replace(to_replace='.', value=0, inplace=True)
# data.currency = data.currency.astype(float)
```

print(data)

```python
ts = pd.Series(data.currency)
ts=ts.astype(float)
ts = ts[ts > 0.0]

y_pos = np.arange(len(data.date))
plt.xticks(y_pos, data.currency)
plt.ylabel('DAVIN')
plt.title('DAVIN')




ts.plot()
plt.show()
```
